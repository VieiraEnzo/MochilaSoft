```python
from readinstance import Instance
from knapsack_solution import Solution
import re
```

```python
# Function to extract all best fixed sets from a given file
def extract_all_fixed_sets(file_path):
    with open(file_path, 'r') as file:
        data = file.read()
        
    # Find all occurrences of 'Best Fixed Set'
    matches = re.findall(r'Best Fixed Set:\s*([\d\s]+)', data)
    
    # Convert each matched string of numbers into a list of integers
    all_fixed_sets = [list(map(int, match.split())) for match in matches]
    
    return all_fixed_sets
```

```python
def get_cost(sol_items) -> int:
    instance = Instance("kpf_2_sum.txt")
    mD = instance.forfeit_cost_pairs_mx()
    cost = 0
    
    
    for item in sol_items:
        cost += instance.profits[item]

    for i in sol_items:
        for j in sol_items:
            if i < j: 
                cost = cost - mD[i][j]
    
                
    return cost
```

```python
# Path to the text file
file_path = 'patterns_mfss.txt'

# Get the best fixed set
patterns = extract_all_fixed_sets(file_path)
```

```python
for i in range(len(patterns)):
    print(len(patterns[i]),get_cost(patterns[i]))
```

```python
# Function to read the patterns from the file
def read_patterns(file_path):
    patterns = []
    
    with open(file_path, 'r') as file:
        for line in file:
            # Check if the line starts with "pattern" and extract numbers
            if line.startswith("pattern"):
                # Extract the part after the colon and convert to a list of integers
                numbers = list(map(int, line.split(':')[1].strip().split()))
                patterns.append(numbers)
    
    return patterns

# Usage example
file_path = 'patterns_hdmils.txt'  # Replace with your actual file path
patterns_hdmils = read_patterns(file_path)
```

```python
# Function to extract all fixed sets from the .txt file and calculate costs
def extract_fixed_sets_and_calculate_cost(file_path):
    with open(file_path, 'r') as file:
        data = file.read()

    # Find all occurrences of lists in square brackets
    matches = re.findall(r'\[([\d,\s]+)\]', data)
    
    # Convert each match (which is a string) into a list of integers
    all_fixed_sets = []
    for match in matches:
        # Split by comma and remove any empty strings or spaces
        cleaned_numbers = [num for num in match.split(',') if num.strip()]
        # Convert the cleaned numbers to integers
        fixed_set = list(map(int, cleaned_numbers))
        all_fixed_sets.append(fixed_set)
    
    # Calculate cost for each list using the cost function
    all_costs = [get_cost(fixed_set) for fixed_set in all_fixed_sets]
    
    return all_fixed_sets, all_costs
```

```python
fixed_sets, costs = extract_fixed_sets_and_calculate_cost('solution_b_solver.txt')
```

```python
# Output the fixed sets and their corresponding costs
for i, (fixed_set, cost) in enumerate(zip(fixed_sets, costs)):
    print(f"Fixed Set {i+1}: {fixed_set}")
    print(f"Cost: {cost}\n")
```
